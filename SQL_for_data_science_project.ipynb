```
#importing pandas
import pandas as pd
#loading sql extension
%load_ext sql
#connecting to sql lite database
%sql sqlite://
```

```
#import and preview of the dataset
accidents=pd.read_csv('traffic_accidents.csv')
accidents.head()
```

```
#delete the table if it exists in our database
%sql DROP TABLE if EXISTS accidents;
#Create table in our database
%sql PERSIST accidents;
```

```
# view first five rows in the table 
%sql SELECT * FROM accidents LIMIT 5;
```

**Display a list of 5 makes that had the highest number of crashes**

```
#Display a list of 5 makes that had the highest number of crashes
%%sql 
SELECT make,COUNT(make) as accident_frequency   FROM accidents 
GROUP BY make 
ORDER BY accident_frequency DESC
LIMIT 5;
```

*Toyota Motor Company* had the highest number of accidents with *637* accidents followed *Chevrolet,Ford,Nissan and Honda* with *573, 566, 491 and  369* accidents respectively

**Most personal vehicles involved in crashes were travelling towards which direction?**

```
#Most personal vehicles involved in crashes were travelling towards which direction? 
#
%%sql
SELECT  travel_direction ,COUNT(vehicle_use) as personal_vehicles_frequency FROM accidents
WHERE vehicle_use='PERSONAL'
GROUP BY travel_direction
ORDER BY personal_vehicles_frequency DESC;
```

personal vehicles heading *North* were involved in most accidents with a total of *885* accidents 

**How many passengers were involved in personal cars were in accidents?**

```
#How many passengers were involved in personal cars were in accidents?
%%sql
SELECT  SUM(num_passengers) as total_passangers FROM accidents
WHERE vehicle_use='PERSONAL';
```

A total of *869*  passangers were in accidents involving personal cars

**How many passengers were involved in accidents while travelling in personal Sport Utility vehicles?**

```
# How many passengers were involved in accidents while travelling in personal Sport Utility vehicles?
%%sql
SELECT  SUM(num_passengers) as total_passangers FROM accidents
WHERE vehicle_type='SPORT UTILITY VEHICLE (SUV)';
```

A total of *174*  passangers were in accidents involving *sport utility vehicles*

**Compute the total number of Mercedez-Benz vehicles involved in accidents while being parked.**

```
#Compute the total number of Mercedez-Benz vehicles involved in accidents while being parked.
%%sql
SELECT COUNT(make) as no_of_mercedez_accidents_parked FROM accidents
WHERE make='MERCEDES-BENZ' and maneuver='PARKED'
```

*10 mercedez benz vehicles* were involved in accidents while *parked*

**Compute the total occupant count per vehicle count.**

```
#Compute the total occupant count per vehicle count.
%%sql
SELECT make,AVG(occupant_cnt) as Total_occupant_count_per_vehicle_count 
FROM accidents
GROUP BY make
ORDER BY Total_occupant_count_per_vehicle_count DESC;
```

 **Sort the number of passengers in descending order by travel direction taking into account travel direction. Return travel direction and no. of passengers.**

```
%%sql
SELECT travel_direction, SUM(num_passengers) as no_of_passangers from accidents
GROUP BY travel_direction
ORDER BY no_of_passangers DESC;
```

Majority of the passangers involved in accidents were headed *west* with a total of *297* passangers followed by *North, East, South, South East, South West ,Unkown destination, North East and North West* with *295, 283, 283, 23, 18, 14, 8, 8* respectively

**Select top 5 vehicle models involved in the accident with no defect, order them by make and show their manoeuvre.** 

```
%%sql
SELECT make,model,maneuver,COUNT(make) as no_of_vehicles FROM accidents
WHERE vehicle_defect= 'NONE'
GROUP BY model
ORDER BY no_of_vehicles DESC
LIMIT 5;
```

Most of the vehicles involved in accident were of *mercedez benz* make unknown model with *405 vehicles* followed by *Toyota Camry, Nissan Altima,Nissan Sentra and Chevrolet*  with *135, 82, 72, 69* respectively 

**Which ridesharing vehicles in an accident had the most number of passengers? Return model, and no of passengers**

```
%%sql
SELECT model,SUM(num_passengers) as No_of_passangers FROM accidents
WHERE vehicle_use='RIDESHARE SERVICE'
GROUP BY model
ORDER BY No_of_passangers DESC
LIMIT 5;
```

*Corolla and MKZ model* from the rideshare service had the most number of passangers with 5 passangers each followed by *Camry, Prius and unknown models* with *4* passangers each
